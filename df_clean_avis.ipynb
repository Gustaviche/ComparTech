```python
import pandas as pd
import re
```

```python
df = pd.read_csv('text_translate.csv', index_col=0)
```

```python
pd.set_option('max_colwidth',None)
```

```python
pd.reset_option('max_colwidth')
```

```python
df.drop(columns=['Unnamed: 0'], inplace=True)
```

```python
# Regex pour extraire le pays
regex = r'(?:Avis laissé en|Commenté en|Avis laissé au|Avis laissé aux) ([^\s]+)'

# Fonction pour extraire le pays
def extract_country(date):
    match = re.search(regex, date)
    if match:
        return match.group(1)
    return None

# Appliquer la fonction à la colonne 'Date' et créer une nouvelle colonne 'Pays'
df['Pays'] = df['Date'].apply(extract_country)
```

```python
# Regex pour extraire la date
regex_date = r'\b\d{1,2} \w+ \d{4}\b'

# Fonction pour extraire la date
def extract_date(text):
    match = re.search(regex_date, text)
    if match:
        return match.group(0)  # Utiliser group(0) car il n'y a pas de parenthèses capturantes
    return None

# Appliquer la fonction à la colonne 'Date' et créer une nouvelle colonne 'Date complète'
df['Date complète'] = df['Date'].apply(extract_date)

df
```

```python
df.dropna(subset=['Avis'], inplace=True)
```

```python
df['Étoiles'] = df['Étoiles'].astype(int)
```

```python
df.drop(columns=['Lien_Avis'], inplace=True)
```
